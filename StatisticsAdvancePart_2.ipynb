Q1 - What is hypothesis testing in statistics?

Hypothesis testing is a fundamental concept in statistics used to make inferences or draw conclusions about a population based on a sample of data. It helps to determine whether there is enough evidence to support a particular claim or hypothesis about the population.

Q2 -  What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H₀) is a statement that there is no effect, difference, or relationship in a population. It represents the default or status quo assumption that researchers aim to test against.

The alternative hypothesis (H₁ or Ha) is a statement that contradicts the null hypothesis. It suggests that there is an effect, difference, or relationship in the population.

Key difference:

The null hypothesis assumes no change or effect.
The alternative hypothesis suggests that there is a change or effect.
In hypothesis testing, we test whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

Q3 -  What is the significance level in hypothesis testing, and why is it important?

The significance level (denoted as α) is the probability threshold set by the researcher before conducting a hypothesis test. It represents the likelihood of making a Type I error, which occurs when the null hypothesis is incorrectly rejected when it is actually true. Commonly used values for α are 0.05, 0.01, and 0.10.

Importance:
It determines how strong the evidence must be before rejecting the null hypothesis.
A smaller α (e.g., 0.01) makes it harder to reject the null hypothesis, reducing the risk of a false positive, while a larger α (e.g., 0.10) increases the risk.
In essence, the significance level helps control the balance between avoiding false positives and detecting real effects.

Q4 - What does a P-value represent in hypothesis testing?

In hypothesis testing, the P-value represents the probability of obtaining a test statistic at least as extreme as the one observed, assuming that the null hypothesis (H₀) is true.

Interpretation:
A small P-value (typically ≤ 0.05) suggests that the observed data is unlikely under H₀, leading to rejection of the null hypothesis in favor of the alternative hypothesis (H₁).
A large P-value (> 0.05) suggests that the observed data is consistent with H₀, meaning there is insufficient evidence to reject it.

Q5 - How do you interpret the P-value in hypothesis testing?

The P-value in hypothesis testing is interpreted as the probability of observing data as extreme as (or more extreme than) the sample result, assuming the null hypothesis (H₀) is true. It helps determine whether there is enough evidence to reject H₀.

Interpretation of P-values:
Small P-value (≤ α, typically 0.05)

Strong evidence against H₀.
Reject the null hypothesis.
Suggests the alternative hypothesis (H₁) is more likely.
Large P-value (> α)

Weak or no evidence against H₀.
Fail to reject the null hypothesis.
The data is consistent with H₀.
Example Interpretation:
P-value = 0.02 (less than 0.05):
→ The result is statistically significant; reject H₀.
P-value = 0.08 (greater than 0.05):
→ The result is not statistically significant; fail to reject H₀.

Q6 - What are Type 1 and Type 2 errors in hypothesis testing?

In hypothesis testing, Type I and Type II errors occur when incorrect conclusions are drawn about the null hypothesis (H₀).

1. Type I Error (False Positive)
Occurs when H₀ is true, but we incorrectly reject it.
It is denoted by α (significance level), typically set at 0.05 (5%).
Example: A medical test wrongly detects a disease in a healthy person.

2. Type II Error (False Negative)
Occurs when H₀ is false, but we fail to reject it (i.e., we incorrectly accept it).
It is denoted by β, and 1 - β is the test’s power.
Example: A medical test fails to detect a disease in a sick person.

Q7 - What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

The key difference between a one-tailed and a two-tailed test in hypothesis testing is how the alternative hypothesis (H₁) is formulated and where we check for significance in the distribution.

1. One-Tailed Test (Directional Test)
Tests for an effect in only one direction (either greater than or less than a value).
Used when we have a specific expectation about the direction of the effect.

Significance Region: Entire alpha (α) is placed in one tail of the distribution.
Example: Checking if a new drug increases recovery speed (H₁: Mean recovery time is lower).

2. Two-Tailed Test (Non-Directional Test)
Tests for an effect in both directions (greater than or less than a value).
Used when we are checking for any significant difference, without specifying direction.

Significance Region: Alpha (α) is split equally between both tails.
Example: Checking if a new teaching method affects student scores (it could increase or decrease scores).

Q8 - What is the Z-test, and when is it used in hypothesis testing?

What is a Z-test?
A Z-test is a statistical test used to determine whether two population means are different when the population variance is known and the sample size is large (n≥30). It follows the standard normal distribution (Z-distribution).

When is a Z-test used?
A Z-test is appropriate when:

The sample size is large (n≥30), ensuring the Central Limit Theorem applies.
The population variance is known (or the sample variance can approximate it well).
The data is approximately normally distributed, especially for small samples.

Q9 - How do you calculate the Z-score, and what does it represent in hypothesis testing?

The Z-score (or standard score) measures how many standard deviations a data point or sample mean is from the population mean. In hypothesis testing, the Z-score helps determine whether to reject the null hypothesis (H₀) by comparing it to critical values from the standard normal distribution.

What Does the Z-score Represent in Hypothesis Testing?
The Z-score indicates how far the sample mean is from the population mean in standard deviation units.
It helps determine whether the observed result is statistically significant.

Q10 - What is the T-distribution, and when should it be used instead of the normal distribution?

The T-distribution (Student’s t-distribution) is a probability distribution similar to the normal distribution but has heavier tails. It is used when estimating population parameters with small sample sizes (n<30) or when the population standard deviation (σ) is unknown.

When to Use the T-Distribution Instead of the Normal Distribution?
Use the T-distribution when:

The sample size is small (n<30).
The population standard deviation (σ) is unknown, so the sample standard deviation (s) is used instead.
The data is approximately normally distributed (or for large n, the Central Limit Theorem helps).

Q11 - What is the difference between a Z-test and a T-test?

A Z-test is used when the sample size is large (n≥30) and the population standard deviation (σ) is known. It follows the normal distribution and is generally used when population parameters are well-defined. On the other hand, a T-test is used when the sample size is small (n<30) and the population standard deviation is unknown, so the sample standard deviation (s) is used instead. The T-test follows the T-distribution, which has heavier tails to account for increased variability in small samples. While both tests compare means to determine statistical significance, the Z-test is more precise for larger datasets, whereas the T-test is better suited for small sample sizes with higher uncertainty.

Q12 -  What is the T-test, and how is it used in hypothesis testing?

A T-test is a statistical test used to determine whether there is a significant difference between the means of two groups, or if a sample mean differs significantly from a known population mean. It is used when the sample size is small (n<30) or when the population standard deviation (σ) is unknown, and the sample standard deviation (s) is used instead.

A researcher wants to test if the average weight of a sample of 25 apples is different from the known average weight of apples (200g). The sample mean weight is 195g, and the sample standard deviation is 10g. A one-sample T-test can be used to compare the sample mean to the population mean and determine if the difference is statistically significant.

Q13 - What is the relationship between Z-test and T-test in hypothesis testing?

Both tests are used for testing the mean of a population, but the Z-test is used when the population standard deviation is known or for large samples, whereas the T-test is used when the population standard deviation is unknown or for small samples.
As the sample size grows, the T-distribution approaches the normal distribution, meaning the results of the Z-test and T-test converge for large samples.

Q14 - What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate an unknown population parameter (e.g., mean, proportion). It provides a range of plausible values for the parameter, with a certain level of confidence (e.g., 95%).

How is it used?
The CI is constructed around a sample statistic (e.g., sample mean).
It is accompanied by a confidence level (e.g., 95%), which represents the probability that the interval contains the true population parameter if the experiment were repeated multiple times.
A wider CI indicates more uncertainty about the estimate, while a narrower CI suggests greater precision.

Q15 - What is the margin of error, and how does it affect the confidence interval?

The margin of error (MoE) is the amount of error allowed in a statistical estimate. It reflects the range above and below the sample statistic (e.g., sample mean) within which the true population parameter is expected to lie, based on a chosen confidence level (e.g., 95%).

How does it affect the confidence interval?
The margin of error directly affects the width of the confidence interval (CI).
A larger margin of error results in a wider CI, indicating more uncertainty.
A smaller margin of error results in a narrower CI, indicating more precision.

Q16 - How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem is a fundamental concept in probability theory and statistics that allows us to update the probability of a hypothesis based on new evidence. It provides a way to calculate the posterior probability of an event, given prior knowledge and new data.

Helps update beliefs: It allows us to revise our beliefs or hypotheses in light of new data.
Decision making: In many fields (e.g., medical diagnosis, spam filtering, machine learning), Bayes' Theorem is used to make more informed decisions by incorporating both prior knowledge and new evidence.
Flexibility: It’s particularly useful when dealing with uncertainty and incomplete information.

Q17 - What is the Chi-square distribution, and when is it used?

The Chi-square distribution is a probability distribution used for analyzing categorical data. It is mainly used in two key tests:

1)Goodness-of-Fit Test: Checks if a sample matches a population distribution.

2)Test for Independence: Assesses if two categorical variables are independent.

It is defined by degrees of freedom (df) and is used to calculate a Chi-square statistic, which compares observed and expected frequencies. If the statistic exceeds a critical value, the null hypothesis is rejected.

Q18 - What is the Chi-square goodness of fit test, and how is it applied?

The Chi-square goodness of fit test is used to determine whether a set of observed categorical data matches an expected distribution. It compares the observed frequencies of categories with the expected frequencies under the assumption that the null hypothesis is true.

Application in Short:
Null Hypothesis (H₀): The observed data fits the expected distribution.
Alternative Hypothesis (H₁): The observed data does not fit the expected distribution.

Q19 - What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is a probability distribution that arises when comparing two variances. It is the ratio of two scaled chi-square distributions and is used primarily in variance analysis. The F-distribution is right-skewed and has two parameters: degrees of freedom for the numerator and denominator.

When is it Used in Hypothesis Testing?

1)Analysis of Variance (ANOVA):
Used to compare the means of multiple groups by analyzing the variances within and between the groups.
Example: Testing if the average test scores differ across three different teaching methods.

2)F-test for Equality of Variances:
Used to compare the variances of two populations to see if they are equal.
Example: Comparing the variances of two groups' exam scores to test if their variability is the same.

Q20 - What is an ANOVA test, and what are its assumptions?

The Analysis of Variance (ANOVA) test is a statistical method used to compare the means of three or more groups to determine if there is a significant difference between them. It tests the null hypothesis that all group means are equal.

1)Independence: The samples from different groups must be independent of each other.

2)Normality: The data in each group should follow a normal distribution (especially important for small sample sizes).

3)Homogeneity of Variances (Equal Variances): The variance within each group should be approximately equal.

Q21 - What are the different types of ANOVA tests?

Here are the main types of ANOVA tests:

One-Way ANOVA: Compares means of three or more independent groups based on one factor (e.g., different teaching methods).

Two-Way ANOVA: Compares means based on two independent factors and tests their interaction (e.g., teaching method and study time).

Repeated Measures ANOVA: Used when the same subjects are measured multiple times (e.g., measuring the same participants before, during, and after treatment).

MANOVA: Compares multiple dependent variables simultaneously (e.g., testing teaching method effects on both scores and attendance).

ANCOVA: Combines ANOVA and regression to compare means while controlling for covariates (e.g., adjusting for baseline knowledge).

Q22 - What is the F-test, and how does it relate to hypothesis testing?

The F-test is a statistical test used to compare two variances or to test if the variances of multiple groups are significantly different. It is often used in ANOVA (Analysis of Variance) to determine if the means of different groups are significantly different by comparing the variability between and within groups.

F-test in Hypothesis Testing:

Null Hypothesis (H₀): The variances are equal (for comparing variances) or the means are equal (for ANOVA).

Alternative Hypothesis (H₁): The variances are not equal (for comparing variances) or the means are not equal (for ANOVA).

Q1 - Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy import stats

# Given data
population_mean = 100  # Known population mean
sample_data = [98, 102, 101, 97, 100, 99, 104, 105, 98, 100]  # Sample data
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Multiply by 2 for two-tailed test

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

Q4 - Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Given data
population_mean = 100  # Known population mean
sample_data = np.random.normal(loc=population_mean, scale=15, size=50)  # Simulating a sample
sample_size = len(sample_data)

# Calculate sample mean and sample standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)

# Perform the Z-test
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Multiply by 2 for two-tailed test

# Output the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Set significance level
alpha = 0.05

# Visualization of the decision region
x = np.linspace(-4, 4, 1000)  # Range for Z-scores
y = stats.norm.pdf(x, loc=0, scale=1)  # Standard normal distribution

# Plotting the normal distribution curve
plt.plot(x, y, label="Standard Normal Distribution", color='blue')

# Shade the rejection regions
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.5, label="Rejection Region")

# Mark the Z-score on the plot
plt.axvline(x=z_score, color='green', linestyle='--', label=f'Z-Score: {z_score:.2f}')

# Add labels and title
plt.title('Two-Tailed Z-Test with Decision Region')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()

# Show plot
plt.show()

# Hypothesis testing interpretation
if p_value < alpha:
    print("\nReject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

Q5 - Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function to calculate and visualize Type I and Type II errors
def plot_type_errors(population_mean, sample_size, population_std, effect_size, alpha=0.05):
    # Step 1: Calculate the critical value (Z-critical) for Type I error (alpha)
    z_critical = stats.norm.ppf(1 - alpha / 2)  # For two-tailed test

    # Step 2: Calculate the rejection region
    rejection_region_left = population_mean - z_critical * (population_std / np.sqrt(sample_size))
    rejection_region_right = population_mean + z_critical * (population_std / np.sqrt(sample_size))

    # Step 3: Generate sample data for both the true population (H0) and the alternative hypothesis (H1)
    sample_data_h0 = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_data_h1 = np.random.normal(loc=population_mean + effect_size, scale=population_std, size=sample_size)

    # Step 4: Perform Z-test to calculate Type I and Type II errors
    sample_mean_h0 = np.mean(sample_data_h0)
    sample_mean_h1 = np.mean(sample_data_h1)

    # Z-scores for both cases (null hypothesis and alternative hypothesis)
    z_score_h0 = (sample_mean_h0 - population_mean) / (population_std / np.sqrt(sample_size))
    z_score_h1 = (sample_mean_h1 - population_mean) / (population_std / np.sqrt(sample_size))

    # Type I error: Probability of rejecting H0 when H0 is true (falls in rejection region)
    type_1_error = 2 * (1 - stats.norm.cdf(z_critical))  # Two-tailed test

    # Type II error: Probability of failing to reject H0 when H1 is true (outside rejection region)
    type_2_error = stats.norm.cdf(z_critical - effect_size / (population_std / np.sqrt(sample_size))) + \
                   (1 - stats.norm.cdf(-z_critical - effect_size / (population_std / np.sqrt(sample_size))))

    # Step 5: Visualization
    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)
    y = stats.norm.pdf(x, loc=population_mean, scale=population_std)

    plt.figure(figsize=(10, 6))

    # Plot the normal distribution under H0
    plt.plot(x, y, label="H0: Population Mean", color="blue")

    # Shading for Type I error (rejection region)
    plt.fill_between(x, 0, y, where=(x <= rejection_region_left) | (x >= rejection_region_right), color="red", alpha=0.5, label="Type I Error (Rejection Region)")

    # Plot the alternative distribution under H1 (effect size added)
    y_h1 = stats.norm.pdf(x, loc=population_mean + effect_size, scale=population_std)
    plt.plot(x, y_h1, label="H1: Alternative Population Mean", color="green", linestyle="--")

    # Shading for Type II error (failure to reject region)
    plt.fill_between(x, 0, y_h1, where=(x >= rejection_region_left) & (x <= rejection_region_right), color="yellow", alpha=0.5, label="Type II Error (Failure to Reject H0)")

    # Add labels and title
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.axvline(rejection_region_left, color='red', linestyle='--', label=f'Critical Value: {rejection_region_left:.2f}')
    plt.axvline(rejection_region_right, color='red', linestyle='--')
    plt.legend()

    plt.show()

    # Print out the errors
    print(f"Type I Error (alpha): {type_1_error:.4f}")
    print(f"Type II Error (beta) with effect size {effect_size}: {type_2_error:.4f}")

# Example Usage
population_mean = 100  # Population mean
population_std = 15    # Population standard deviation
sample_size = 30       # Sample size
effect_size = 5        # Effect size (difference between population mean and alternative hypothesis mean)

# Call the function to visualize and calculate Type I and Type II errors
plot_type_errors(population_mean, sample_size, population_std, effect_size)
```

Q6 - Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

# Simulating data for two independent groups
group1 = np.random.normal(loc=75, scale=10, size=30)  # Group 1: Mean = 75, SD = 10, n = 30
group2 = np.random.normal(loc=80, scale=15, size=30)  # Group 2: Mean = 80, SD = 15, n = 30

# Perform the Independent T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Output the results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between the means of the two groups.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the means of the two groups.")
```

Q7 - Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulating paired data (before and after treatment)
before_treatment = np.random.normal(loc=75, scale=10, size=30)  # Data before treatment
after_treatment = before_treatment + np.random.normal(loc=5, scale=5, size=30)  # Data after treatment (mean increase = 5)

# Perform the Paired Sample T-test
t_statistic, p_value = stats.ttest_rel(before_treatment, after_treatment)

# Output the results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between before and after treatment.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between before and after treatment.")

# Visualization
plt.figure(figsize=(8, 6))

# Plotting the before and after treatment data
plt.plot(before_treatment, label="Before Treatment", color='blue', marker='o', linestyle='None')
plt.plot(after_treatment, label="After Treatment", color='green', marker='o', linestyle='None')

# Connecting lines between before and after treatment for each subject
for i in range(len(before_treatment)):
    plt.plot([i, i], [before_treatment[i], after_treatment[i]], color='gray', linestyle='-', alpha=0.5)

plt.title('Comparison of Before and After Treatment (Paired Sample T-test)')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
from scipy import stats

# Function to calculate the confidence interval for a sample mean
def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    sample_size = len(sample_data)

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the critical value (t-distribution)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    return confidence_interval, sample_mean, margin_of_error

# Example usage
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Simulated sample data

confidence_interval, sample_mean, margin_of_error = calculate_confidence_interval(sample_data)

# Output the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

Q16 - Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy import stats

# Simulating two random samples (e.g., sample 1 and sample 2)
np.random.seed(42)
sample1 = np.random.normal(loc=100, scale=15, size=50)  # Sample 1: Mean = 100, SD = 15
sample2 = np.random.normal(loc=100, scale=25, size=50)  # Sample 2: Mean = 100, SD = 25

# Calculate sample variances
var1 = np.var(sample1, ddof=1)  # Sample variance (ddof=1 for sample variance)
var2 = np.var(sample2, ddof=1)

# Perform the F-test
f_statistic = var1 / var2 if var1 > var2 else var2 / var1  # Ensure the larger variance is in the numerator
dof1 = len(sample1) - 1  # Degrees of freedom for sample 1
dof2 = len(sample2) - 1  # Degrees of freedom for sample 2

# Calculate the p-value from the F-distribution
p_value = 1 - stats.f.cdf(f_statistic, dof1, dof2)

# Output the results
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")
```

Q19 - Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to check normality, independence, and equal variance assumptions for ANOVA
def check_anova_assumptions(groups):
    # 1. Check Normality using Shapiro-Wilk test for each group
    print("Checking normality assumption:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk Test: Statistic={stat:.4f}, P-value={p_value:.4f}")
        if p_value < 0.05:
            print("    The data in this group is not normally distributed (reject H₀).")
        else:
            print("    The data in this group is normally distributed (fail to reject H₀).")

    # 2. Check Homogeneity of Variance using Levene's test
    print("\nChecking homogeneity of variance (equal variance) assumption:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test: Statistic={stat:.4f}, P-value={p_value:.4f}")
    if p_value < 0.05:
        print("    The variances are significantly different (reject H₀).")
    else:
        print("    The variances are not significantly different (fail to reject H₀).")

    # 3. Checking for independence: Independence is assumed by study design.
    print("\nIndependence assumption:")
    print("    Independence is assumed by the design of the experiment, so no formal test is performed.")

    # 4. Visualize the data distribution using boxplots
    print("\nVisualizing the groups using boxplots:")
    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title('Boxplot of Groups')
    plt.xlabel('Group')
    plt.ylabel('Values')
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1: Mean = 50, SD = 10
group2 = np.random.normal(loc=55, scale=15, size=30)  # Group 2: Mean = 55, SD = 15
group3 = np.random.normal(loc=60, scale=20, size=30)  # Group 3: Mean = 60, SD = 20

# List of groups
groups = [group1, group2, group3]

# Check the assumptions
check_anova_assumptions(groups)
```

Q24 - Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy import stats

# Simulating sample data
np.random.seed(42)

# Sample size (n) and hypothesized population variance (sigma^2)
n = 30
hypothesized_variance = 100  # Hypothesized population variance (sigma^2)

# Generate random data from a normal distribution with sample variance
sample_data = np.random.normal(loc=0, scale=10, size=n)  # Standard deviation = 10, variance = 100
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for sample variance)

# Chi-square test statistic
chi_square_statistic = (n - 1) * sample_variance / hypothesized_variance

# Degrees of freedom
df = n - 1

# Calculate the p-value (two-tailed test)
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

# Output the results
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The sample variance is significantly different from the hypothesized population variance.")
else:
    print("\nFail to reject the null hypothesis: The sample variance is not significantly different from the hypothesized population variance.")
```

Q25 - Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from scipy.stats import norm

# Function to perform Z-test for proportions
def z_test_proportions(x1, n1, x2, n2):
    # Sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled sample proportion
    p_pool = (x1 + x2) / (n1 + n2)

    # Z-test statistic formula
    z = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z)))

    return z, p_value

# Simulated data for two groups (successes and sample sizes)
x1 = 120  # Number of successes in group 1
n1 = 300  # Sample size of group 1
x2 = 100  # Number of successes in group 2
n2 = 250  # Sample size of group 2

# Perform the Z-test for proportions
z_statistic, p_value = z_test_proportions(x1, n1, x2, n2)

# Output the results
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The proportions are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The proportions are not significantly different.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Simulate two datasets with different variances
np.random.seed(42)

# Dataset 1: Mean = 50, SD = 10, Sample size = 30
dataset_1 = np.random.normal(loc=50, scale=10, size=30)

# Dataset 2: Mean = 50, SD = 15, Sample size = 30
dataset_2 = np.random.normal(loc=50, scale=15, size=30)

# Calculate the sample variances
var1 = np.var(dataset_1, ddof=1)  # Sample variance (ddof=1 for sample variance)
var2 = np.var(dataset_2, ddof=1)

# Perform the F-test (F-statistic = var1 / var2)
f_statistic = var1 / var2 if var1 > var2 else var2 / var1  # Ensure the larger variance is in the numerator

# Degrees of freedom
df1 = len(dataset_1) - 1  # Degrees of freedom for dataset 1
df2 = len(dataset_2) - 1  # Degrees of freedom for dataset 2

# Calculate the p-value (two-tailed test)
p_value = 1 - f.cdf(f_statistic, df1, df2)

# Output the results
print(f"Sample Variance 1: {var1:.4f}")
print(f"Sample Variance 2: {var2:.4f}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The variances of the two datasets are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances of the two datasets are not significantly different.")

# Visualization of the datasets and their distributions
plt.figure(figsize=(10, 6))

# Plot histograms of the two datasets
plt.hist(dataset_1, bins=10, alpha=0.7, label='Dataset 1 (SD=10)', color='blue')
plt.hist(dataset_2, bins=10, alpha=0.7, label='Dataset 2 (SD=15)', color='red')

plt.title('Comparison of Variances between Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

Q27 -  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulating categorical data (e.g., results of a dice roll)
# We simulate a sample of 100 rolls of a fair six-sided die
observed_data = np.random.randint(1, 7, size=100)

# Calculate the observed frequencies for each side (1-6)
observed_freq = [np.sum(observed_data == i) for i in range(1, 7)]

# Define the expected frequencies for a fair die (each side should appear equally)
expected_freq = [100 / 6] * 6  # Since it's a fair die, each outcome should appear with equal probability

# Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(observed_freq, expected_freq)

# Output the results
print(f"Observed Frequencies: {observed_freq}")
print(f"Expected Frequencies: {expected_freq}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fits the expected distribution.")

# Visualization of the observed vs expected frequencies
labels = [1, 2, 3, 4, 5, 6]

plt.bar(labels, observed_freq, alpha=0.7, label='Observed', color='blue', width=0.4, align='center')
plt.bar(labels, expected_freq, alpha=0.7, label='Expected', color='red', width=0.4, align='edge')
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.title('Chi-square Goodness-of-Fit Test: Observed vs Expected Frequencies')
plt.legend()
plt.grid(True)
plt.show()
```
